# Home Court Advantage

Goal: **Does home court advantage exist? If it does exist, how impactful is it in correlation to an ELO system and predicting a winner?**

```python
import pandas as pd
import numpy as np
from pathlib import Path
from datetime import datetime
import requests
import bs4
import sqlite3
```

```python
conn = sqlite3.connect('..' / Path('data') / 'nba.db')
df = pd.read_sql("""
    SELECT
        date            AS "Date",
        start_time      AS "Start Time (ET)",
        visitor         AS "Visitor",
        visitor_points  AS "Visitor Points",
        home            AS "Home",
        home_points     AS "Home Points",
        box_score       AS "Box Score",
        overtime        AS "Overtime",
        attendance      AS "Attendance",
        arena           AS "Arena",
        notes           AS "Notes"
    FROM games ORDER BY date DESC
                 
""", conn)
conn.close()
df
```
